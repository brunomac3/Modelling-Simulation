```python
import gymnasium as gym
import highway_env
from stable_baselines3 import DQN

# 1. Create the environment (no rendering for faster training)
env = gym.make('highway-v0')

# 2. Define the RL model
# We'll use DQN with a Multi-Layer Perceptron policy (MlpPolicy)
model = DQN('MlpPolicy', env, 
            learning_rate=5e-4, 
            buffer_size=15000, 
            batch_size=32,
            gamma=0.8,
            verbose=1) # verbose=1 prints training progress

# 3. Train the model
# 20,000 timesteps is a very short run, just for this test.
# A real run would be 1e5 or 1e6.
print("Starting training...")
model.learn(total_timesteps=20000)
print("Training complete!")

# 4. Save the trained model
model.save("dqn_highway_agent")
print("Model saved!")
```
